```python
# aggregate function using list
genres.agg(['mean','max','std','sum'])
```

```python
# aggregate function using both list and dictonary
genres.agg(
    {
       'IMDB_Rating':['mean','sum'],
        'Runtime':['max','mean'],
        'Gross':['sum','max'],
        'Metascore':['std','sum']
    }
)
```

```python
# loops

for group,data in genres:
    print(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()])
```

```python
df = pd.DataFrame(columns = movies.columns)

for group,data in genres:
    df = df.append(print(data[data['IMDB_Rating'] == data['IMDB_Rating'].max()]))
df
```

```python
# apply function

genres.apply(min)
```

```python
# find no.of movies starting with A for eack group

def foo(group):
    return group['Series_Title'].str.startswith('A').sum()
genres.apply(foo)
```

```python
# find the rank of each movies iun the group according to the imdb score

def rank_movies(group):
    group['genre_rank'] = group['IMDB_Rating'].rank(ascending = False)

    return group
genres.apply(rank_movies)
    
```

```python
# find normalized IMDP rating group wise

def normal(group):
    group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group
genres.apply(normal)
```
