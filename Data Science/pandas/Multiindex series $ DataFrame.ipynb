```python
# pivot table
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips.groupby('sex').mean()
```

```python
tips.groupby('sex')[['total_bill']].mean()
```

```python
tips.groupby(['sex','smoker'])[['total_bill']].mean()
```

```python
tips.pivot_table(index='sex',columns='smoker',values='total_bill')
```

```python
tips.groupby(['sex','smoker'])[['total_bill']].mean().unstack()
```

```python
# aggfunc
tips.pivot_table(index='sex',columns='smoker',values='total_bill',aggfunc='std')
```

```python
# multidimensional
tips.pivot_table(index=['sex','smoker'],columns=['day','time'],aggfunc={'size':'mean','tip':'max','total_bill':'sum'},margins=True)
```
