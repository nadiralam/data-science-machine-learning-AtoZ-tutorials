```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('sharma-kohli.csv')
df
```

```python
plt.plot(df['index'],df['RG Sharma'])
plt.plot(df['index'],df['V Kohli'])
```

```python
plt.plot(df['index'],df['RG Sharma'],color = 'Black',linestyle = 'solid',linewidth = 3,marker = 'D',markersize = 10,label = 'rohit')
plt.plot(df['index'],df['V Kohli'],color = 'Green',linestyle = 'solid',linewidth = 3,marker = 'D',markersize = 10,label = 'virat')
plt.xlabel('season')
plt.ylabel('runs')
plt.title('comparision between kolhi and rohit')
plt.legend()
plt.grid()
plt.show()
```
