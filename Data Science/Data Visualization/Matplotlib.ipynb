```python
# labels and title

plt.plot(roko['index'],roko['V Kohli'],)
plt.plot(roko['index'],roko['RG Sharma'])
plt.title('Comparision Between V Kohli and RG Sharma')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
# colors(hex) and line(width and style) and marker(size)

plt.plot(roko['index'],roko['V Kohli'],color = 'Green')
plt.plot(roko['index'],roko['RG Sharma'],color = 'Black')
plt.title('Comparision Between V Kohli and RG Sharma')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
# linestyle and linewidth 

plt.plot(roko['index'],roko['V Kohli'],color = 'Green',linestyle = 'solid',linewidth = 3)
plt.plot(roko['index'],roko['RG Sharma'],color = 'Black',linestyle = 'dashdot', linewidth = 2)
plt.title('Comparision Between V Kohli and RG Sharma')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
# marker

plt.plot(roko['index'],roko['V Kohli'],color = 'Green',linestyle = 'solid',linewidth = 3,marker = 'D')
plt.plot(roko['index'],roko['RG Sharma'],color = 'Black',linestyle = 'dashdot', linewidth = 2,marker = 'o')
plt.title('Comparision Between V Kohli and RG Sharma')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
# legend location

plt.plot(roko['index'],roko['V Kohli'],color = 'Green',linestyle = 'solid',linewidth = 3,marker = 'D',label = 'Virat')
plt.plot(roko['index'],roko['RG Sharma'],color = 'Black',linestyle = 'dashdot', linewidth = 2,marker = 'o',label = 'Rohit')
plt.title('Comparision Between V Kohli and RG Sharma')
plt.xlabel('season')
plt.ylabel('runs scored')
plt.legend(loc = 'upper right')
```

```python
# markersize

plt.plot(roko['index'],roko['V Kohli'],color = 'Green',linestyle = 'solid',linewidth = 3,marker = 'D',label = 'Virat',markersize = 10)
plt.plot(roko['index'],roko['RG Sharma'],color = 'Black',linestyle = 'dashdot', linewidth = 2,marker = 'o',label = 'Rohit',markersize = 5)
plt.title('Comparision Between V Kohli and RG Sharma')
plt.xlabel('season')
plt.ylabel('runs scored')
plt.legend()
```

```python
# bar chart

children = [10,40,30,20,50]
colour = ['red','black','green','blue','indigo']
plt.bar(colour,children)
```

```python
# color, label etc
plt.bar(colour,children, color = 'black',label = 'colour_liking')
plt.title('Colour liking by Students')
plt.xlabel('COlour')
plt.ylabel('Children')
plt.legend()
```

```python
# barh-----> using for horizantal graph plot 
plt.barh(colour,children, color = 'black',label = 'colour_liking')
plt.title('Colour liking by Students')
plt.xlabel('COlour')
plt.ylabel('Children')
plt.legend()
```

```python
# multiple bar graph

plt.bar(np.arange(df1.shape[0]) - 0.2,df1['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df1.shape[0]),df1['2016'],width=0.2,color='red')
plt.bar(np.arange(df1.shape[0]) + 0.2,df1['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df1.shape[0]), df1['batsman'])

plt.show()
```

```python
# Stacked Bar chart
plt.bar(df1['batsman'],df1['2017'],label='2017')
plt.bar(df1['batsman'],df1['2016'],bottom=df1['2017'],label='2016')
plt.bar(df1['batsman'],df1['2015'],bottom=(df1['2016'] + df1['2017']),label='2015')

plt.legend()
plt.show()
```
