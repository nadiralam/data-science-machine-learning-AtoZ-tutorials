```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
tip = sns.load_dataset("tips")
tip
```

```python
# it's a axes level function
sns.scatterplot(data = tip, x = 'total_bill',y = 'tip',color  = 'red',marker = '+',hue = 'sex',style = 'time',size = 'size')
```

```python
# relplot is a function level function
sns.relplot(data = tip,x  = 'total_bill',y = 'tip',kind = 'scatter',color = 'red',style = 'time',size = 'size',hue = 'sex')
```

```python
# line plot
sns.relplot(data = tip,x  = 'total_bill',y = 'tip',kind= 'line')
```

```python
# kdeplot
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate
sns.kdeplot(data=tip,x='total_bill')
```

```python
sns.displot(data=tip,x='total_bill',kind='kde')
```

```python
# hue -> fill
sns.displot(data=tip,x='total_bill',kind='kde',hue='sex',fill=True,height=10,aspect=2)
```

```python

sns.displot(data=tip,x='total_bill',kind='kde',hue='sex',height=10,aspect=2)
```

```python
# Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.kdeplot(data=tip,x='total_bill')
sns.rugplot(data=tip,x='total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot 
# and then shows the count of observations within each rectangle with the fill color

# sns.histplot(data=tips, x='total_bill', y='tip')
sns.displot(data=tip, x='total_bill', y='tip',kind='hist')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data=tip, x='total_bill', y='tip')
```

```python
# matrix plot---> Heatmap, Clustermap

# Matrix plot

fig = gap.pivot(index = 'country',columns = 'year',values = 'lifeExp')
plt.figure(figsize = (15,7))
sns.heatmap(fig)
```
