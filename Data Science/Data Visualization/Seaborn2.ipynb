# Plotting Pairwise Relationship (PairGrid Vs Pairplot)

```python
sns.pairplot(data = df1)
```

```python
df1
```

```python
# hue

sns.pairplot(data = df1,hue = 'Species')
```

```python
n = sns.PairGrid(data = df1,hue = 'Species')
n.map_diag(sns.boxplot)
n.map_offdiag(sns.kdeplot)
```

```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(data=df1,hue='Species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# vars
g = sns.PairGrid(data=df1,hue='Species',vars=['SepalWidthCm','PetalWidthCm'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

# JointGrid Vs Jointplot

```python
sns.jointplot(data = df,x = 'tip',y = 'total_bill',hue = 'sex',kind = 'hist')
```

```python
g = sns.JointGrid(data=df,x='total_bill',y='tip')
g.plot(sns.kdeplot,sns.violinplot)
```
