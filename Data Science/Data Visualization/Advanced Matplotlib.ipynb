```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris  = pd.read_csv("iris.csv")
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.sample(5)
```

```python
# Scatter plot

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c = iris['Species'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
# cmap and colorbar and alpha

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c = iris['Species'],cmap = 'winter',alpha = 0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
# Size

plt.figure(figsize = (5,3))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c = iris['Species'],cmap = 'winter',alpha = 0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
# Annotation

batter = pd.read_csv("batter.csv")
batter.head(5)
```

```python
batter_df = batter.head(100).sample(25)
batter_df
```

```python
plt.figure(figsize = (18,10))

plt.scatter(batter_df['avg'],batter_df['strike_rate'],color = 'red')

for i in range(batter_df.shape[0]):
    plt.text(batter_df['avg'].values[i],batter_df['strike_rate'].values[i],batter_df['batter'].values[i])
```

```python
# size

plt.figure(figsize = (18,10))

plt.scatter(batter_df['avg'],batter_df['strike_rate'],color = 'red',s = batter_df['runs'])

for i in range(batter_df.shape[0]):
    plt.text(batter_df['avg'].values[i],batter_df['strike_rate'].values[i],batter_df['batter'].values[i])
```

```python
# Draw lines for analysing the batter stats

plt.figure(figsize = (18,10))

plt.scatter(batter_df['avg'],batter_df['strike_rate'],color = 'red',s=batter_df['runs'])
plt.axhline(130,color = 'red')
plt.axvline(30,color = 'red')

for i in range(batter_df.shape[0]):
    plt.text(batter_df['avg'].values[i],batter_df['strike_rate'].values[i],batter_df['batter'].values[i])
```
